### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
fh = open("./data/64_72hr_CH.xml", "r", encoding="utf8")
xml = fh.read()
fh.close()

# print(xml)
```

```python
import xmltodict
d = dict(xmltodict.parse(xml))
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
localnames = d['cwbopendata']['dataset']['locations']['location']
```

```python
# 1. 請問高雄市有多少地區有溫度資料？

'''
Your Code
'''
print("Total: ", len(localnames))
for local in localnames:
    print(local['locationName'])
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

'''
Your Code
'''

for local in localnames:
    print(local['locationName'])
#     print(local['weatherElement'][0]["time"][0]['dataTime'])
    for weather in local['weatherElement']:
        if weather['description'] == '溫度':
            print(weather["time"][0]['dataTime'])
            print("{}: {}".format(weather['description'], weather["time"][0]['elementValue']['value'] ))
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

'''
Your Code
'''
local = localnames[0]

for weather in local['weatherElement']:  
    if weather['description'] == '溫度':
        for t in range(len(weather["time"])):
            print(weather["time"][t]['dataTime'])
            print("溫度: ", weather["time"][t]['elementValue']['value'])
#         if weather["time"][t]["elementValue"]["measures"] = = '攝氏度':
#             print(weather["time"][t]['dataTime'])
#             print(weather["time"][t]['elementValue']['value'])
```
